```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('bjlianjia.xlsx', header = None)
```

```python
df.head()
```

```python
columns = ["房型","成交时间","地址","价格","朝向","装修","建造年代","楼层","产权","面积","得房面积","是都有钥匙","编号"]
df.columns = columns
```

```python
df.head(20)
```

## check na

```python
df.isnull().sum()
```

```python
df.columns
```

## drop na

```python
df = df.dropna(subset = ['房型','成交时间','地址','朝向','装修','建造年代','楼层', '产权', '面积', '得房面积','是都有钥匙', '编号'], how ='any')
```

```python
df.isnull().sum()
```

## 戶型處理

```python
df[['室','厅','厨','卫']] = df['房型'].str.extract('(\d+)室(\d+)厅(\d+)厨(\d+)卫')
df.drop('房型', axis = 1 , inplace = True)
```

## 成交時間處理

```python
df['deal_time'] = df['成交时间'].map(lambda x : x.split()[0])
df['deal_from'] = df['成交时间'].map(lambda x : x.split()[1])
```

```python
df.drop(columns = '成交时间', inplace = True)
```

```python
df.rename(index = str, columns = {'deal_time':'成交時間', 'deal_from':'成交自'}, inplace = True)
```

```python
df['成交時間'] = pd.to_datetime(df['成交時間'], format = '%Y.%m.%d')
```

## 價格處理

```python
#先split樓層，然後groupby樓層去分配平均價格

df['floor'] = df['楼层'].map(lambda x : x.split()[0])
df['total_floor'] = df['楼层'].map(lambda x : x.split()[1])
```

```python
df.groupby('floor')['价格'].mean()
```

```python
df['价格'] = df['价格'].fillna(df.groupby('floor')['价格'].transform('mean'))
```

```python
df.drop('楼层', axis = 1 , inplace = True)
```

```python
df.rename(index = str, columns = {'floor':'樓層','total_floor':'總樓層' }, inplace = True)
```
